#### 4. 모델 시각화

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()
X = iris.data

# PCA 적용 (3개의 주성분 추출)
pca = PCA(n_components=3)
x_pca = pca.fit_transform(X)

# PCA 결과를 3D로 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', elev=48, azim=134)  # Axes3D 객체 생성

# 3D 스캐터 플롯 (PCA의 3개의 주성분)
ax.scatter(x_pca[:, 0], x_pca[:, 1], x_pca[:, 2], c=iris.target, edgecolor='w', s=100)

# 축 레이블 설정 및 레이블 제거
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# plot 크기 설정
ax.dist = 12

plt.show()
```

#### (Extra) 5. 몇 개의 PC 면 충분할까?

```python
# 각각의 새로운 축이 데이터셋의 분산(variance)을 얼마나 표현하는지 확인이 가능
# [PC1, PC2, PC3, ...]

# Hint: Explain variance
model.explained_variance_ratio_
```

```python
# 몇 개의 PC면 충분할까?
# np.argmax : 최대값의 인덱스를 리턴
# np.cumsum : 누적된 합계를 계산

np.argmax(np.cumsum(model.explained_variance_ratio_) >= 0.95 ) + 1
# 95% 이상의 variance 를 설명하기 위한 축의 갯수를 확인할 수 있음
```

```python
np.argmax(np.cumsum(model.explained_variance_ratio_) >= 0.95 ) + 1
```

```python
# Better option (indicate the ratio of variance you wish to preserve)

iris = datasets.load_iris()
x = iris.data
y = iris.target

model = decomposition.PCA(n_components=0.95) # Explains 95% of total variance
model.fit(x)
x = model.transform(x)
x
```
